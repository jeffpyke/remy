```python
# reproducing figure 2 from the paper

import matplotlib.pyplot as plt
import matplotlib

legend = ['2x', '10x', '100x', '1000x', '1000x-updated']
files = ['data-2014-2x.dna.5.csv', 'data-2014-10x.dna.5.csv', 'data-2014.dna.4.csv', 'data-2014-1000x.dna.4.csv', 'data-2014-1000x.dna.2.csv']
assert(len(legend) == len(files))
X = []
Y = []

link_speeds = set()

for f in files:
    fd = open(f, 'r')
    lines = fd.readlines()
    x = []
    y = []
    for line in lines:
        l = line.split(',')
        link_speed = float(l[0])
        link_speeds.add(link_speed)
        x.append(10*link_speed)
        
        score = float(l[1])
        y.append(score)
    X.append(x)
    Y.append(y)

for i in range(len(files)):
    plt.semilogx(X[i], Y[i], label=legend[i])
plt.axvspan(22, 44, facecolor='b', alpha=0.25)
plt.axvspan(10, 100, facecolor='y', alpha=0.15)
plt.axvspan(3.2, 320, facecolor='g', alpha=0.2)
plt.axis([1, 1000, -2, 0])
plt.yticks([0, -1, -2, -3])
plt.legend(loc=9, mode="expand", ncol=len(legend))
plt.xlabel('Link Speed(Mbps)')
plt.ylabel('Normalized score')
plt.grid(True)
plt.savefig('range_reproduction.png')
plt.show()
```
